```python
# A program that compares all the DNA sequences in a fasta file and print the pairwise percentage identity of the 
# sequences (# of matches/the total length). For example, a sample fasta file shown will output:
# sequencex vs sequencey 90%
# sequencex vs sequencez 100%
# sequencey vs sequencez 90% 
# Note: (1). the file format should be always header + sequence. Header always begins with “>”, followed by the sequence 
# name or ID, in one line, then the sequence will begin in a separate line, can be in multiple lines, e.g. 
# >sequencex
# atcg
# cga
# tta
# Different DNA sequences can have difference lengths. For % identity, it’ll be the number of positions that are the 
# same between two DNAs, the total length would be the longer sequence’s length
```

```python
# creating another fasta file

from random import randint

# Creating functions for the random DNA sequences
def random_base(RNAflag = False):
    return ('UCAG' if RNAflag else 'TCAG')[randint(0,3)]

def random_nuc(RNAflag = False):
    return random_base(RNAflag)+random_base(RNAflag)+random_base(RNAflag)

def random_nucs(minlength, maxlength, RNAflag):
    return [random_nuc(RNAflag = False) for n in range(randint(minlength, maxlength))]

outfile = open('anotherout.fas', 'w')
for i in range(1,6):
    dna =random_nucs(7,10, RNAflag= False)
    print ('>dna'+ str(i), file = outfile)
    print(''.join([str(x) for x in dna]), file = outfile)

outfile.close()
```
